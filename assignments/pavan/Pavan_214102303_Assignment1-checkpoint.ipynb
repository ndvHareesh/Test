#### Unit Step Signal
* $$ x_{1}(t) = u(t) =\begin{cases} 
    0 & \text{$t$ <0}\\ 
    1 & \text{0 $\leq$ $t$ $\leq20$ }\\ 
    \end{cases} $$
    
### Triangular Signal
$$ x_{5}(t) = rect(t/tau)*rect(t/tau)  $$    
  
  
#### Triangular Signal
* $$ x_{7}(t) =\begin{cases} 
    0 & \text{$t$ <0}\\ 
    (t-d1)/(d2-d1) & \text{d1 $\leq$ $t$ $\leq d2$ }\\
    (d3-t)/(d3-d2) & \text{d2 $\leq$ $t$ $\leq d3$ }\\
    0 & \text{$t$ >15}\\ 
    \end{cases} $$

```python
#1.x1(t) Unit step signal
def u(a,c,t): #U(at+c)
    if a*t+c>=0:
        return 1
    else:
        return 0
    
def r(a,c,t): #r(at+c)
    if a*t+c>=0:
        return a*t+c
    else:
        return 0
    

def p(a,c,t): #p(at+c)
    if a*t+c>=0:
        return (a*t)**2/2+c
    else:
        return 0

def rect(a,c,t,T): #Rect(at+c)
    if (-0.5*T-c)/a<=t<=(0.5*T-c)/a:
        return 1
    else:
        return 0
    
def tri(a,c,t,T):
    if (-T-c)/a<=t<=(T-c)/a:
        return (1 -(abs(t+c/a)/T))
    else:
        return 0
    
    
def x7(d1,d2,d3,t):
    if t<=d1:
        return 0
    elif d1<=t<=d2:
        return (t-d1)/(d2-d1)
    elif d2<=t<=d3:
        return (d3-t)/(d3-d2)
    elif d3<=t:
        return 0
    
t=np.linspace(-20,20,500)
Y1=[u(1,0,x) for x in t]

Y2=[r(1,0,x) for x in t]

Y3=[p(1,0,x) for x in t]

Y4=[rect(1,0,x,4) for x in t]

Y5=[tri(1,0,x,4) for x in t]

Y6=np.sin(np.pi*t)/(np.pi*t)

Y7=[x7(3,10,15,x) for x in t]

fig, ax = plt.subplots(3, 3, figsize=(25,15))
ax[0][0].plot(t,Y1,'b')
ax[0][0].set_title('Unit Step Signal',size=20)
ax[0][0].set_ylabel('U(t)')

ax[0][1].plot(t,Y2,'g')
ax[0][1].set_title('Ramp signal',size=20)
ax[0][1].set_ylabel('r(t)')

ax[0][2].plot(t,Y3,'r')
ax[0][2].set_title('Parabolic Signal',size=20)
ax[0][2].set_ylabel('p(t)')

ax[1][0].plot(t,Y4,'c')
ax[1][0].set_title('Rectangular Pulse',size=20)
ax[1][0].set_ylabel('rect(t/T)')

ax[1][1].plot(t,Y5,'m')
ax[1][1].set_title('rect(t/T)*rect(t/T)',size=20)

ax[1][1].set_title('Convolution of rectangular pulses',size=20)

ax[1][2].plot(t,Y6,'y')
ax[1][2].set_title('Sinc Function',size=20)

ax[2][0].plot(t,Y7,'k')
ax[2][0].set_title('x7(t)',size=20)
fig.delaxes(ax[2][1])
fig.delaxes(ax[2][2])
plt.tight_layout()
plt.show()
```

#### In continuous-time domain, the impulse function, unit step function, ramp function, and parabolic function are related by the differentiation operation. That is if we differentiate ramp function we get unit step function and so on. Is there any such relation in the discrete-time domain? If yes, how will you implement ramp function from parabolic function, unit step function from ramp function, and impulse function from unit step function in the discrete-time domain

##### Ans: In discrete time domain
$$ \delta[n] = u[n]-u[n-1] $$
$$ u[n] = r[n]-r[n-1] $$
$$ r[n] = p[n]-p[n-1] $$

```python
def un(n):
    u=[]
    for i in n:
        if i>=0:
            u.append(1)
        else:
            u.append(0)
    return u
def rn(n):
    r=[]
    for i in n:
        if i>=0:
            r.append(i)
        else:
            r.append(0)
    return r
def pn(n):
    p=[]
    for i in n:
        if i>=0:
            p.append(i**2/2)
        else:
            p.append(0)
    return p

n=np.arange(-10,10,1)

#Implementing ramp from parabolic signal
p=pn(n)
R=[] #ramp signal
for i in range(len(p)):
    R.append((p[i]-p[i-1]))
    
#Implementing unit step signal from ramp signal
r=rn(n)
U=[] #unit step signal
for i in range(len(r)):
    U.append((r[i]-r[i-1]))

    #Implementing impulse signal from unit step function
u=un(n)
I=[] #Impulse signal
for i in range(len(u)):
    I.append((u[i]-u[i-1]))

functions=[I,U,R]
titles=["Impulse signal","Unit step signal","Ramp signal"]
fig = plt.figure(10,figsize=(25,15))
n=np.arange(-9,10,1)
k=1
for title,f in zip(titles,functions):
    plt.subplot(3,1,k)
    k+=1
    plt.stem(n,f[1:])
    plt.title(title,size=20)
plt.tight_layout()
```

## LTI Systems
### Let x[n] be the input to an linear-time invariant (LTI) system and h[n] be the impulse response. The output of a discrete LTI system in terms of the impulse response is characterized by the equation as:
$$ y[n] = x[n]*h[n] = \sum_{k=-\infty}^{\infty}\ x[k]h[n-k] $$

```python
def convolution(inputsignal,impulseresponse):
        A=np.size(inputsignal) #length of input signal x[n]
        B=np.size(impulseresponse) #length of impulse response h[n]
        o = np.zeros(A+B-1)
        for a in range(A):
            for b in range(B):
                o[a+b] = o[a+b] + inputsignal[a]*impulseresponse[b]
        return o 
```

```python
xn=[3, 5, 2, 6, 8]
hn=[-2,4,1]
yn=convolution(xn,hn)
print('Output sequence = ',yn)
fig=plt.figure(1,figsize=(25,10))
plt.subplot(3,1,1)
plt.stem(xn)
plt.title("Input x[n]",size=20)
plt.subplot(3,1,2)
plt.stem(hn)
plt.title("Impulse Response h[n]",size=20)
plt.subplot(3,1,3)
plt.stem(yn)
plt.title("Output y[n]",size=20)
plt.tight_layout()
plt.show()
```

```python
#h=[0,3] #impulse response delta(n-1)
n=np.arange(-5,5,1)
def h(n):
    h=[]
    for i in n:
        if i==1:
            h.append(3)
        else:
            h.append(0)
    return h

exp1=[np.exp(2+3*j)*i for i in n]
#exp1 = np.array([cmath.exp((2+3j)*n) for i in n ])
y1 = convolution(h(n),un(n))#u[n] * h[n]
y2 = convolution(h(n),rn(n))#r[n] * h[n]
y3 = convolution(h(n),pn(n))#p[n] * h[n]
y4 = convolution(h(n),exp1)

n=np.arange(-9,10,1)
plt.figure(figsize=(18,10))
plt.subplot(5,1,1)
plt.stem(n,y1)
plt.title('Output for unit step')

plt.subplot(5,1,2)
plt.stem(n,y2)
plt.title('Output for ramp')

plt.subplot(5,1,3)
plt.stem(n,y3)
plt.title('output for parabolic')

plt.subplot(5,1,4)
plt.stem(n,y4.real)
plt.title('output of real part for exponetial input')

plt.subplot(5,1,5)
plt.stem(n,y4.imag)
plt.title('output of img part for exponetial input')

plt.tight_layout()
```

#### Consider an input x[n] and a unit impulse response h[n] given by
$$ x[n] = (1/3)^{-n} u[n  1] $$
$$ h[n] = u[n  1] $$
#### Determine and plot the output y[n] = x[n]  h[n]

```python
def u(a,b,n): #u(an+b)
    k=[]
    for i in n:
        if a*i+b>=0:
            k.append(1)
        else:
            k.append(0)
    return np.array(k)
n=np.arange(-10,10,1)
xn=((1/3)**-n)*u(-1,-1,n) #x[n]=(1/3)^-n * u(-n-1)
hn=u(1,-1,n) #h[n]=u[n-1]
yn=convolution(xn,hn)
print('The output of x[n]*h[n] is y[n]=',yn)
fig=plt.figure(1,figsize=(25,10))
plt.subplot(3,1,1)
plt.stem(n,xn)
plt.title("Input x[n]",size=20)
plt.subplot(3,1,2)
plt.stem(n,hn)
plt.title("Impulse Response h[n]",size=20)
plt.subplot(3,1,3)
k=np.arange(-19,20,1)
plt.stem(k,yn)
plt.title("Output y[n]",size=20)
plt.tight_layout()
plt.show()
```

####  Consider the first-order difference equation: y[n]+2y[n1] = x[n]. Assuming the condition of initial rest (i.e., if x[n] = 0 for n < n0, then y[n] = 0 for n < n0), find the impulse response of a system whose input and output are related by this difference equation

#### $Impulse\ Response\ of \ system$,$ y[n] + 2y[n-1] = x[n]$
##### $h[n] = (-2)^{n-n_0}u[n-n_0]$

```python
#impulse signal
def impulse(x):
    return 1 * (x == 0)

n = np.arange(-10,20,1)
x1n=impulse(n)
y = np.zeros(len(n))
for i in range(len(n)):
    y[i+1] = x1n[i+1] - 2*y[i]
    if i==27:
        break
plt.stem(n,y)
plt.title('h[n]')
plt.show()
```
